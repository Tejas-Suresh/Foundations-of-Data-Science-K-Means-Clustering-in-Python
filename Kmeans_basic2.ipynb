```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=pd.read_csv("BankDataset.csv")
f.head()
```

```python
print(np.mean(f))
print(np.std(f))
np.median(f)
```

```python
#indicating deviations in data..since there isn't much deviation from the mean I can say the data is strong.
print(np.mean(f)-(np.std(f)*0.25))
print(np.mean(f)+(np.std(f)*0.25))
```

```python
plt.scatter(f['V1'],f['V2'])
```

```python
from sklearn.cluster import KMeans
ar=np.column_stack((f['V1'],f['V2']))
kmres=KMeans(n_clusters=2).fit(ar)
kmres
```

```python
col=kmres.cluster_centers_
ar=[np.mean(f['V1']),np.mean(f['V2'])]
arstd=[np.std(f['V1']),np.std(f['V2'])]
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(f['V1'],f['V2'])
plt.scatter(ar[0],ar[1],color="yellow",s=400)
plt.scatter(arstd[0],arstd[1],color="orange",s=400)
plt.scatter(col[:,0],col[:,1],s=3000,color="red",alpha=0.25)
```

4 clusters have been plotted to properly classify the given data. I felt 4 clusters was the most stable. The yellow dot represents the mean of the given data and the orange dot its standard deviation.
